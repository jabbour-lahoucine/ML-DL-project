```python

def plot_confusion_matrix(y_test,predictions):
    acc = round(accuracy_score(y_test,predictions), 2)
    cm = confusion_matrix(y_test,predictions)
    sns.heatmap(cm, annot=True, fmt=".0f")
    plt.xlabel('y_pred')
    plt.ylabel('y')
    plt.title('Accuracy Score: {0}'.format(acc), size=10)
    plt.show()
plot_confusion_matrix(y_test, predictions)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix

def generate_test_cases(n_samples=100):
    """
    Generate test cases for diabetes prediction
    """
    np.random.seed(42)

    test_data = {
        'Pregnancies': np.random.randint(0, 17, size=n_samples),
        'Glucose': np.random.normal(loc=120.89, scale=31.97, size=n_samples).astype(int),
        'BloodPressure': np.random.normal(loc=69.10, scale=19.36, size=n_samples).astype(int),
        'SkinThickness': np.random.normal(loc=20.54, scale=15.95, size=n_samples).astype(int),
        'Insulin': np.random.exponential(scale=79.80, size=n_samples).astype(int),
        'BMI': np.random.normal(loc=31.99, scale=7.88, size=n_samples),
        'DiabetesPedigreeFunction': np.random.exponential(scale=0.47, size=n_samples),
        'Age': np.random.normal(loc=33.24, scale=11.76, size=n_samples).astype(int)
    }

    # Clip values to realistic ranges
    test_data['Glucose'] = np.clip(test_data['Glucose'], 0, 200)
    test_data['BloodPressure'] = np.clip(test_data['BloodPressure'], 0, 122)
    test_data['SkinThickness'] = np.clip(test_data['SkinThickness'], 0, 99)
    test_data['Insulin'] = np.clip(test_data['Insulin'], 0, 846)
    test_data['BMI'] = np.clip(test_data['BMI'], 0, 67.1)
    test_data['DiabetesPedigreeFunction'] = np.clip(test_data['DiabetesPedigreeFunction'], 0.078, 2.42)
    test_data['Age'] = np.clip(test_data['Age'], 21, 81)

    return pd.DataFrame(test_data)

# Generate test cases
X_test = generate_test_cases(100)

# Assuming your trained model is called 'model'
# Replace 'model' with your actual model variable name
y_pred = log.predict(X_test)

# Example test cases for individual predictions
test_cases = [
    # Low risk case
    [1, 85, 66, 29, 0, 26.6, 0.351, 31],
    # High risk case
    [8, 183, 72, 35, 125, 34.5, 0.672, 55],
    # Moderate case
    [3, 130, 70, 25, 80, 29.4, 0.315, 40]
]

print("Individual Test Case Predictions:")
print("--------------------------------")
for case in test_cases:
    features = pd.DataFrame([case], columns=X_test.columns)
    prediction = log.predict(features)[0]
    probability = log.predict_proba(features)[0][1]

    print(f"\nTest Case:")
    for feature, value in zip(X_test.columns, case):
        print(f"{feature}: {value}")
    print(f"Prediction: {'Diabetic' if prediction == 1 else 'Non-diabetic'}")
    print(f"Probability of diabetes: {probability:.2%}")
```

```python
import pandas as pd

def predict_diabetes(model, pregnancies, glucose, blood_pressure, skin_thickness,
                    insulin, bmi, diabetes_pedigree, age):
    """
    Make diabetes prediction for custom input values
    """
    # Create a single row dataframe with the input values
    input_data = pd.DataFrame([[pregnancies, glucose, blood_pressure, skin_thickness,
                               insulin, bmi, diabetes_pedigree, age]],
                             columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                                    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])

    # Make prediction
    prediction = model.predict(input_data)[0]
    probability = model.predict_proba(input_data)[0][1]

    return prediction, probability

# Example usage (replace with your values):
pregnancies = 6
glucose = 148
blood_pressure = 72
skin_thickness = 35
insulin = 0
bmi = 33.6
diabetes_pedigree = 0.627
age = 50

prediction, probability = predict_diabetes(log, pregnancies, glucose, blood_pressure,
skin_thickness, insulin, bmi, diabetes_pedigree, age)

print(f"Prediction: {'Diabetic' if prediction == 1 else 'Non-diabetic'}")
print(f"Probability of diabetes: {probability:.2%}")
```
